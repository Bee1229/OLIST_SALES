```python
#converting date columns
#ORDERITEMS TABLE


order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])
order_items['shipping_limit_date'] = order_items['shipping_limit_date'].dt.date
print(order_items.info())


#Order review table
data_col1 = ['review_creation_date', 'review_answer_timestamp']

for i in data_col1:
    order_reviews[i] = pd.to_datetime(order_reviews[i]).dt.date

print(order_reviews.info())
#ORDERS TABLE

data_col2 = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

for i in data_col2:
    orders[i] = pd.to_datetime(orders[i], format='mixed')
    
print(orders.info())
```

**FEATURE ENGINEERING**

```python
#DELIVERY TIME

orders['delivery_time'] = orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']

orders['delivery_time'] = orders['delivery_time'].dt.days
orders.head(5)
```

```python
#Estimated delivery day

orders['estimated_delivery_day'] = orders['order_estimated_delivery_date'] - orders['order_purchase_timestamp']

orders['estimated_delivery_day'] = orders['estimated_delivery_day'].dt.days
```

```python
#Convert both to datetime

orders['delivery_time'] = pd.to_datetime(orders['delivery_time'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
#late delivery 
# we will assume late delivery is any oder that arrived more than 7 days later

orders['late_delivery'] = orders['delivery_time'] > orders['order_estimated_delivery_date']


orders.head(5)
```

```python
orders['delivery_time'] = (
    orders['order_delivered_customer_date'] 
    - orders['order_purchase_timestamp']
).dt.days
```
